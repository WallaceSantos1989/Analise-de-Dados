```python
def fix_col_name(cols: list) -> list:
  '''
  Corrije o nome das colunas eliminando caracteres indesejados, bem como
  espaços, parêntesis etc...

  Params:
    cols - recebe a lista de colunas do dataframe

  Return:
    Retorna uma lista com os nomes tratados
  '''
  col_list_tmp = []
  for col in cols:
    col_tmp_name = col[col.index(',')+3:-2]
    col_list_tmp.append(col_tmp_name)
  return col_list_tmp
```

#Confirmando as hipóteses

#1. Escolas em áreas de maior renda per capita apresentam notas mais altas, o desempenho varia regionalmente com alguma região se destacando?

```python
# Agrupando por 'SG_UF_ESCOLA' e calculando a média das notas
media_total_por_estado = df.groupby('SG_UF_ESCOLA')['NU_MEDIA_TOT'].mean().reset_index()

# Arredondando a média para duas casas decimais
media_total_por_estado['media_total'] = media_total_por_estado['NU_MEDIA_TOT'].round(2)

# Mapear os estados para suas respectivas regiões
# Exemplo de mapeamento (adapte conforme necessário)
regioes = {
    'AC': 'Norte', 'AL': 'Nordeste', 'AP': 'Norte', 'AM': 'Norte',
    'BA': 'Nordeste', 'CE': 'Nordeste', 'DF': 'Centro-Oeste', 'ES': 'Sudeste',
    'GO': 'Centro-Oeste', 'MA': 'Nordeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste',
    'MG': 'Sudeste', 'PA': 'Norte', 'PB': 'Nordeste', 'PR': 'Sul',
    'PE': 'Nordeste', 'PI': 'Nordeste', 'RJ': 'Sudeste', 'RN': 'Nordeste',
    'RS': 'Sul', 'RO': 'Norte', 'RR': 'Norte', 'SC': 'Sul',
    'SP': 'Sudeste', 'SE': 'Nordeste', 'TO': 'Norte'
}

# Adicionando a coluna de região ao DataFrame
media_total_por_estado['Regiao'] = media_total_por_estado['SG_UF_ESCOLA'].map(regioes)

# Agrupando por região e calculando a média das notas
media_total_por_regiao = media_total_por_estado.groupby('Regiao')['media_total'].mean().reset_index()

# Criando um gráfico de barras para visualizar a média por região
plt.figure(figsize=(10, 6))
sns.barplot(data=media_total_por_regiao, x='Regiao', y='media_total', palette='viridis')
plt.title('Média de Notas por Região do Brasil')
plt.xlabel('Região')
plt.ylabel('Média Total')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

De acordo com os dados históricos de 2005 a 2015, conseguimos ver que as maiores médias de fato são na reigão Sul e Sudeste, seguidos de Centro-Oeste, Nordeste e Norte.

#2. Desempenho escolar melhorou de 2005 a 2015 devido a investimentos em educação?

```python
# Filtrando e agrupando os dados
media_taxa_aprovacao = (
    df[df['NU_TAXA_APROVACAO'].notnull()]  # Filtra valores não nulos
    .groupby('NU_ANO')['NU_TAXA_APROVACAO']  # Agrupa por ano
    .mean()  # Calcula a média
    .reset_index()  # Redefine o índice
)

# Arredondando a média para duas casas decimais
media_taxa_aprovacao['media_taxa_aprovacao'] = media_taxa_aprovacao['NU_TAXA_APROVACAO'].round(2)

# Exibindo a tabela de evolução
print("Evolução da Média da Taxa de Aprovação:")
print(media_taxa_aprovacao)

# Criando o gráfico de evolução da taxa de aprovação
plt.figure(figsize=(10, 6))
sns.lineplot(data=media_taxa_aprovacao, x='NU_ANO', y='media_taxa_aprovacao', marker='o')
plt.title('Evolução da Média da Taxa de Aprovação do ENEM ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Média da Taxa de Aprovação (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid()
plt.show()
```

Podemos notar de forma expressiva o aumento na taxa de aprovação de acordo com os anos, o que confirma a hipótese referente aos investimentos na educação dentro do período estudado. Segue abaixo uma notícia que confirma essa hipótese.


Analisando o período 2001-2015, o estudo compara indicadores das distribuições de renda original com aqueles decorrentes do investimento em educação.

No período 2001 a 2005, a progressividade do investimento em educação pública cresceu e se estabilizou, enquanto a contribuição da educação à renda ficou estável. No período 2005-2015, a progressividade manteve-se em patamar alto, enquanto cresceu a contribuição da educação à renda.

O esforço do governo federal com educação refletiu-se nos diversos níveis de ensino, mas com intensidade três vezes maior na educação básica do que no ensino superior. Em valores correntes (dos respectivos anos), o investimento por estudante na educação básica passou de  899,00 reais em 2000 para 7.273,00 reais em 2015, enquanto o da educação superior sobe de 8.849,00 reais para 23.215,00 reais.

Em 2001, o investimento por estudante em educação superior era mais de 10 vezes (10,47) superior do que o mesmo valor destinado para cada estudante da educação básica. Em 2015, essa diferença caiu para cerca de 3 vezes (3,64).

Fonte: https://campanha.org.br/noticias/2020/10/16/investimento-publico-em-educacao-teve-efeitos-redistributivos-na-renda-do-brasileiro-mostra-estudo/

#3. Redes escolares administradas por entidades públicas possuem um desempenho melhor?

```python
# Mapeando os tipos de dependência administrativa
dependencia_mapping = {
    1: 'Federal',
    2: 'Estadual',
    3: 'Municipal',
    4: 'Privada'
}

# Substituindo os códigos pelos nomes correspondentes
df['TP_DEPENDENCIA_ADM_ESCOLA'] = df['TP_DEPENDENCIA_ADM_ESCOLA'].replace(dependencia_mapping)

# Agrupando e calculando a média das notas
media_por_dependencia = (
    df.groupby('TP_DEPENDENCIA_ADM_ESCOLA')['NU_MEDIA_TOT']  # Agrupa por tipo de dependência
    .mean()  # Calcula a média
    .reset_index()  # Redefine o índice
)

# Renomeando a coluna da média
media_por_dependencia.columns = ['Tipo de Dependência', 'Média Total']

# Exibindo a tabela de média por tipo de dependência
print("Média da Nota Total por Tipo de Dependência Administrativa:")
print(media_por_dependencia)

# Criando o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(media_por_dependencia['Média Total'],
        labels=media_por_dependencia['Tipo de Dependência'],
        autopct='%1.1f%%',
        startangle=140)
plt.title('Média Total das Notas')
plt.axis('equal')  # Igualando o aspecto para um gráfico de pizza
plt.show()
```

Temos a média de cada tipo de entidade de que adminstra as escolas e nomeia em federal, estadual, municipal e privada, além de ordenar de acordo com os resultados do melhor desempenho para o menor.

Entendemos, através da análise que as redes de ensino administradas por entidades federais apresentam um melhor desempenho 31,4%, seguidas pelas redes de ensino particulares 27,7%, estaduais 20,7% e, por último, municipais 20,3%.

#4. Menos alunos por turma resultam em um melhor desempenho?

```python
# Agrupando e calculando a média das notas por porte da escola
media_por_porte = (
    df.groupby('PORTE_ESCOLA')['NU_MEDIA_TOT']  # Agrupa por porte da escola
    .mean()  # Calcula a média
    .reset_index()  # Redefine o índice
)

# Renomeando a coluna da média
media_por_porte.columns = ['Porte da Escola', 'Média Total']

# Exibindo a tabela de média por porte da escola
print("Média da Nota Total por Porte da Escola:")
print(media_por_porte)

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(data=media_por_porte, x='Porte da Escola', y='Média Total', palette='viridis')
plt.title('Média da Nota Total por Porte da Escola')
plt.xlabel('Porte da Escola')
plt.ylabel('Média Total das Notas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid()
plt.show()
```

Existe uma relação entre quantidade de alunos por turma e a qualidade do desempenho, segundo a análise feita, quanto menor a quantidade de alunos, melhor é o desempenho no ENEM.
